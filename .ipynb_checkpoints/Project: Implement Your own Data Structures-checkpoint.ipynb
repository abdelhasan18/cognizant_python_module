Step 1: Create the Inventory

Start with an empty dictionary called inventory.
Each key in the dictionary will represent an item name, and the value will be a tuple containing the quantity and price.
Example:

inventory =
{
"apple": (10, 2.5),
"banana": (20, 1.2)
}

Step 2: Add, Remove, and Update Items

Add functionality to:
Add a new item to the inventory (e.g., "mango": (15, 3.0)).
Remove an item from the inventory.
Update the quantity or price of an existing item.

Step 3: Display the Inventory
Write a loop to display all items in the inventory in a friendly format. For example:

Item: apple, Quantity: 10, Price: $2.5
Item: banana, Quantity: 20, Price: $1.2

Step 4: Bonus - Calculate Total Value
Add a feature to calculate and display the total value of the inventory by multiplying the quantity and price of each item.

Example Run:

Welcome to the Inventory Manager!
Current inventory:
Item: apple, Quantity: 10, Price: $2.5
Item: banana, Quantity: 20, Price: $1.2
Adding a new item: mango
Updated inventory:
Item: apple, Quantity: 10, Price: $2.5
Item: banana, Quantity: 20, Price: $1.2
Item: mango, Quantity: 15, Price: $3.0
Total value of inventory: $90.0

```python
inventory = {}
inventory['apple'] = (10, 2.5)
inventory['banana'] = (20, 1.2)

print('Welcome to Inventory Manager!\n')

def retDict(): # general inventory printing function, prints total value as well
    value = 0
    for item in inventory:
        inv = inventory[item]
        value += inv[0] * inv[1]
        print('Item: {}, Quantity: {}, Price: ${}'.format(item, inv[0], inv[1]))
    print('Total value of inventory: ${}\n'.format(value))

def add(item, quan=15, price=3.0): # adds an item, if quantity and price are not defined, defaults to 15 and $3
    inventory[item] = (quan, price)
    print('Added a new item:', item)
    retDict()

def remove(item): # removes an item
    del inventory[item]
    print('Removed an item:', item)
    retDict()

def update(item, quan=-1, price=-1): # updates an item, if no new quantity or price is defined than nothing is changed
    arr = [['quantity', 0], ['price', 0]] # array used to later track if this feature was changed
    if quan > -1:
        inventory[item] = (quan, inventory[item][1]) # tuples aren't updated so much as they are just overwritten
        arr[0][1] = 1
    if price > -1:
        inventory[item] = (inventory[item][0], price)
        arr[1][1] = 1
    output = ''
    if arr[0][1] or arr[1][1]: # if any new value has been defined
        for feature in arr:
            if feature[1]:
                if len(output) > 0: # in the event that both were changed, output message needs conjunction
                    output += ' and '
                output += feature[0]
        output += ' of ' + item
        output = output.capitalize()
    else:
        output = 'Nothing'
    print(output, 'has been updated.')
    retDict()

retDict()
add('mango')
remove('apple')
update('banana')
update('banana', 30, 2.0)
```
